```python
import numpy as np

x = np.array([0.5,0.2,0.1])
w = np.array([0.4,-0.7,0.2])
y_sum = np.dot(x,w)
output = y_sum
print("Weighted sum: ",y_sum)
print("Output: ",output)
```

```python
def biological_neuron(signals, threshold=1.0):
    soma = sum(signals)
    if soma>=threshold:
        return 1
    return 0
signals  = [0.3,0.4,0.5]
print("Neuron Output: ",biological_neuron(signals))
```

```python
#Artificial neuron with wts+activation
def artificial_neuron(inputs,weights,bias=0):
    y_sum = np.dot(inputs,weights)+bias
    return y_sum
x = np.array([1,0,1])
w = np.array([0.5,-0.6,0.2])
print("Summation output: ", artificial_neuron(x,w,bias=0.1))
```

```python
import math
#Identity fun
def identity(x):
    return x
#Step/Threshold fun
def step(x):
    return 1 if x>=0 else 0
#ReLU fun
def threshold(x,theta=0.5):
    return 1 if x>=theta else 0
#ReLU fun
def relu(x):
    return max(0,x)
#Sigmoid fun
def sigmoid(x,k=1):
    return 1/(1+math.exp(-k*x))
#Bipolar Sigmoid
def bipolar_sigmoid(x):
    return (2/(1+math.exp(-x)))
#Hyperbloic Tangent
def tanh_n(x):
    return math.tanh(x)


values = [-2,-1,0,0.5,1,2]
print("Input values: ",values)
print("\n----Identical function----")
for v in values:
    print(v,"->",identity(v))
print("\n----Step/Threshold function----")
for v in values:
    print(v,"->",step(v))
print("\n----ReLU function----")
for v in values:
    print(v,"->",threshold(v))
print("\n----ReLU function----")
for v in values:
    print(v,"->",relu(v))
print("\n----Sigmoid function----")
for v in values:
    print(v,"->",sigmoid(v))
print("\n----Bipolar function----")
for v in values:
    print(v,"->",bipolar_sigmoid(v))
print("\n----Hyperbloic Tangent function----")
for v in values:
    print(v,"->",tanh_n(v))
```
