```python
#Dataset for Clustering Demonstration
import numpy as np
import matplotlib.pyplot as mat
x= np.random.randint(1,100,6)
y = np.random.randint(1,100,6)
print("X: ",x,"Y: ",y)
xy2 = np.array([[x[i],y[i]] for i in range(6)])
print("XY: ",xy2)

#Visualization before clustering
mat.scatter(xy2[:,0],xy2[:,1])
mat.xlabel('X')
mat.ylabel('Y')
mat.show()
```

```python
# K-Means Clustering
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2 ,random_state=54)
labels = kmeans.fit_predict(xy2)

mat.scatter(xy2[:,0],xy2[:,1],c=labels)
mat.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='x')
mat.title('K-Means Clustering')
mat.xlabel('Feature 1')
mat.ylabel('Feature 2')
mat.grid(True)
mat.show()
```
