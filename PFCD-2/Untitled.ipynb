```python
def power(base, exponent):
    if exponent == 1:
        return base
    return base * power(base, exponent - 1)

print(power(3, 4)) 
```

```python
def gcd(x, y):
    if y == 0:
        return x
    return gcd(y, x % y)

print(gcd(48, 18))
```

```python
def increasing_numbers(n, start=1, path=""):
    if len(path) == n:
        print(path)
        return
    for i in range(start, 10):
        increasing_numbers(n, i + 1, path + str(i))

increasing_numbers(2)
```

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
fib(10)
```

```python
import random

def quickselect(arr, k):
    pivot = random.choice(arr)
    left = [x for x in arr if x > pivot]
    mid = [x for x in arr if x == pivot]
    right = [x for x in arr if x < pivot]

    if k <= len(left):
        return quickselect(left, k)
    elif k <= len(left) + len(mid):
        return pivot
    else:
        return quickselect(right, k - len(left) - len(mid))

print(quickselect([3,2,1,5,6,4], 2))
```

```python
import math

def has_antipodal(points):
    angles = [math.atan2(y, x) for x, y in points]
    angles.sort()

    angle_set = set(round(a, 5) for a in angles)

    for a in angles:
        opposite = round((a + math.pi) % (2 * math.pi), 5)
        if opposite in angle_set:
            return True
    return False

points = [(1, 0), (-1, 0), (0, 1), (0, -1)]
print(has_antipodal(points)) 
```

```python
def partition(arr, low, high):
    pivot = arr[low]
    i = low + 1

    for j in range(low + 1, high + 1):
        if arr[j] < pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1

    arr[low], arr[i - 1] = arr[i - 1], arr[low]
    return i - 1

def quick_sort_helper(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort_helper(arr, low, pi - 1)
        quick_sort_helper(arr, pi + 1, high)

def quick_sort(arr):
    quick_sort_helper(arr, 0, len(arr) - 1)

arr = [37, 2, 6, 4, 89, 8, 10, 12, 68, 45]
quick_sort(arr)
print(arr)
```

```python
data = [
    ("Elon Musk", 433.9),
    ("Jeff Bezos", 239.4),
    ("Mark Zuckerberg", 211.8),
    ("Larry Ellison", 204.6),
    ("Bernard Arnault", 181.3),
    ("Larry Page", 161.4)
]

def selection_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if a[j][1] < a[min_idx][1]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a

sorted_data = selection_sort(data)
result = {name: net for name, net in sorted_data}
print(result)
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i]); i += 1
        else:
            result.append(right[j]); j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

print(merge_sort(['apple','orange','banana','grape']))
```

```python
def merge_lists(a, b):
    i = j = 0
    result = []

    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            result.append(a[i]); i += 1
        else:
            result.append(b[j]); j += 1

    result.extend(a[i:])
    result.extend(b[j:])
    return result

print(merge_lists([1,3,5,7], [2,4,6,8]))
```
